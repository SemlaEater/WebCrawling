# requests를 통한 수집과 BeautifulSoup를 이용한 문서 파싱


```python
headers={
    "User-Agent": (
    "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
    "AppleWebKit/537.36 (KHTML, like Gecko) "
    "Chrome/120.0.0.0 Safari/537.36"
    )
}

response = requests.get(url, headers=headers)

#요청 오류 검사
response.raise_for_status()
```

```python
recipes=[]

for i in range(len(soup.select("div.common_sp_caption > div.common_sp_caption_tit.line2"))):
    name_tag=item.select("div.common_sp_caption > div.common_sp_caption_tit.line2")[i]
    img_tag=item.select("div.common_sp_thumb > a > img")[i]
    writer_tag=item.select("div.common_sp_caption > div.common_sp_caption_rv_name > a")[i]
    view_tag=item.select("div.common_sp_caption > div.common_sp_caption_rv > span.common_sp_caption_buyer")[i]

    name=name_tag.get_text(strip=True) if name_tag else None
    img=img_tag['src'] if img_tag else None
    writer=writer_tag.get_text(strip=True) if img_tag else None
    view=view_tag.get_text(strip=True) if view_tag else None
    view=view.strip('조회수 ')
    
    recipes.append({
        "Name":name,
        "Img":img,
        "Writer":writer,
        "View":view
    })
print(recipes)
```
